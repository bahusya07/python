```
#include <stdio.h>
void main()
{
    int a,b;
    char op;
scanf("%d %d",&a,&b);
scanf(" %c",&op);
switch(op)
{
    case '+': printf("%d",a+b);
    break;
    case '-': printf("%d",a-b);
    break;
    case '*': printf("%d",a*b);
    break;
    case '/': printf("%d",a/b);
    break;
    case '%': printf("%d",a%b);
    break;
    default: printf("invalid input");
    break;
}
}
```

```
#include <stdio.h>
void main()
{
    int n,r,s=0,c=0;
    scanf("%d",&n);
    while(n!=0)
    {
        c++;
        r=n%10;
        s=s+r;
        n=n/10;
    }
    printf("%d %d",s,c);
    }
```

```
#palindrome number
#include <stdio.h>
void main()
{
    int t,n,r,s=0;
    scanf("%d",&n);
    t=n;
    while(t!=0)
    {
        r=t%10;
        s=s*10+r;
        t=t/10;
    }
    if(n==s)
    printf("palindrome number");
    else
    printf("not");
    }
```

```
#program to check a armstrong number
#include <stdio.h>
void main()
{
    int t,n,r,s=0;
    scanf("%d",&n);
    t=n;
    while(t!=0)
    {
        r=t%10;
        s=s+r*r*r;
        t=t/10;
    }
    if(n==s)
    printf("armstrong number");
    else
    printf("not");
    }
```

```
#include <stdio.h>
void main()
{
    int t,c,n,r,s=0;
    scanf("%d",&n);
    t=n;
    while(t!=0)
    {
        c++;
        t=t/10;
    }
    t=n;
    while(t!=0)
    {
        r=t%10;
        s=s+pow(r,c);
        t=t/10;
    }
    if(n==s)
    printf("armstrong number");
    else
    printf("not");
    }

```
