- easy to understand
-better reusability
- lengthy program will be divided into sub programs


#syntax:

def name of the function(<parameters>):

statements

return


```
#input:1234
#output:6(2+4)
def evendigit(n):
  s=0
  while n!=0:
     r=n%10
     if  r %2 ==0:
         s =s+r
     n = n//10
  return s
evendigit(1234)
```

```
#input:19528
#output:9
def largest(n):
    temp = 0
    while n!=0:
      r=n%10
      if r>temp:
         temp=r
      n=n//10
    return temp
largest(19528)
```

```
#reverse of given no
#input:123
#output:321
def reverse(n):
    rev=0
    while n!=0:
       r=n%10
       rev=rev*10+r
       n=n//10
    return rev
reverse(123)
```

```
# print the numbers between two limits
#input:11 25
#output:11 12 13......25
def printseries(lb,ub):
    for i in range(lb,ub+1):
        print(i,end=' ')
    return
printseries(11,25)
```
