```
#program to find whether a no is prime no or not
#include <stdio.h>

void main()
{
int n,i,c=0;
scanf("%d",&n);
for(i=1;i<=n;i++)
{
    if(n%i==0)
    c++;
}
if(c==2)
printf("prime number");
else
printf("not");
}
```

```
#fibanocci series of a particular digit
#include <stdio.h>

void main()
{
int n,f1=0,f2=1,f3,i,c;
scanf("%d",&n);
if(n==1)
c=printf("%d",f1);
else
{
for(i=3;i<=n;i++)
{
    f3=f1+f2;
    f1=f2;
f2=f3;
}
c=printf("%d",f2);
}
printf("\n%d",c);
}
```

```
#fibonnoci series
#include <stdio.h>

void main()
{
int n,f1=0,f2=1,f3,i;
scanf("%d",&n);
printf("fibanoci series:");
for(i=0;i<=n;i++)
{
    printf(" %d",f1);
    f3=f1+f2;
    f1=f2;
f2=f3;
}
return 0;
}
```

```
#include<stdio.h>
int main() {
    int i,j,rows;
    printf("Enter number of rows: ");
    scanf("%d", &rows);
    for (i=1; i<=rows; ++i) {
        for (j=1; j<=i; ++j)
        { printf("%d ",j); }
        printf("\n");
    }
    return 0;
}
```
