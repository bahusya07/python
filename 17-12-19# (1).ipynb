```
# function to return the large number from given list
def largenumber(li):
    li.sort()
    return li[-1]
def genericlarge(li,n):
    li.sort()
    return li[-n]
li = [19,1,25,45,76]
print(largenumber(li))
print(genericlarge(li,3))
```

```
def smallnumber(li):
    li.sort()
    return li[0]
def genericsmall(li,n):
    li.sort()
    return li[n]
li = [19,1,25,45,76]
print(smallnumber(li))
print(genericsmall(li,3))
```
